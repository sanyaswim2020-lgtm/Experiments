```python
# Считаем среднее и медиану для всех параметров 
## в качестве эксперимента для параметров есть расчет как через функции numpy, так и через pandas)
print('Средняя длина чашелистика:', np.mean(df.SepalLengthCm)) 
print('Медиана длины чашелистика:', np.median (df.SepalLengthCm))
' '
print ('Средняя ширина чашелистика:' , np.mean(df.SepalWidthCm))
print('Медиана ширина чашелистика:', np.median (df.SepalWidthCm))
''' '''
print ('Средняя длина лепестка:' , np.mean(df.PetalLengthCm))
print('Медиана длины лепестка:', np.median (df.PetalLengthCm))

print ('Средняя ширина лепестка:' , df['PetalWidthCm'].mean())
print('Медиана ширины лепестка:', df['PetalWidthCm'].median())
```
