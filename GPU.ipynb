```python
import torch

# 1. Проверяем, видит ли PyTorch видеокарту
if torch.cuda.is_available():
    print(f"Yes! GPU is available: {torch.cuda.get_device_name(0)}")
    
    # 2. Создаем тензор (матрицу) на CPU (в оперативной памяти)
    x_cpu = torch.rand(1000, 1000)
    print("Матрица создана в RAM (CPU).")
    
    # 3. МАГИЯ: Переносим матрицу в видеопамять (GPU)
    x_gpu = x_cpu.to('cuda') 
    print("Матрица перенесена в VRAM (GPU). Теперь вычисления будут летать!")
    
    # Пример вычисления на GPU
    result = x_gpu @ x_gpu # Перемножение матриц
    print("Вычисление завершено на GPU.")
else:
    print("GPU не найден. Работаем на процессоре :(")
```
