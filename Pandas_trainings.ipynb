```python
import pandas as pd
import matplotlib.pyplot as pl
```

```python
fruit_sales = pd. DataFrame ({'Apples': [35,41], 'Bananas': [41,34]})
print (fruit_sales)
fruits = pd. DataFrame ({'Apples': [30], 'Bananas': [21]})
print (fruits)

display (fruit_sales)
display (fruits)
```

```python
from IPython.display import display, Markdown

display(Markdown("### Продажи за Январь")) # Выведет красивый жирный заголовок
display(fruit_sales)

display(Markdown("---")) # Выведет горизонтальную черту-разделитель

display(Markdown("### Продажи за Февраль"))
display(fruits)

# .style.background_gradient() заливает ячейки цветом в зависимости от значения
# cmap='Viridis' — это цветовая схема (от синего к желтому)
display(fruit_sales.style.background_gradient(cmap='Greens'))
```
